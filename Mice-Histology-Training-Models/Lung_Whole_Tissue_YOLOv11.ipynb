```python
from ultralytics import YOLO

model = YOLO('yolo11m-seg.yaml')
model = YOLO('yolo11m-seg.pt')

# Train the model
train_results = model.train(
    data="/content/drive/MyDrive/PhD/BioViosion Segmentation/Mouse Samples/Lungs/New whole tissue/Trainning YOLOv11 Model/data.yaml",
    epochs=1000,
    patience=100,
    batch=64,
    imgsz=640,
    device=0,
    lr0=0.0001,
    lrf=0.01,
    amp=True,
    project="/content/drive/MyDrive/PhD/BioViosion Segmentation/Mouse Samples/Lungs/New whole tissue/Trainning YOLOv11 Model/results",  # Explicitly save results here
    name="exp",  # YOLO will create 'exp', 'exp2', etc., automatically
    exist_ok=False  # Set True if you want to overwrite existing results
)
```
