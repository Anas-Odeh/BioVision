```python
import zipfile

zip_file_path = "/content/drive/MyDrive/PhD/BioViosion Segmentation/Mouse Samples/Tongue H&E/Tongue.v1i.yolov11.zip"
extract_location = "/content/drive/MyDrive/PhD/BioViosion Segmentation/Mouse Samples/Tongue H&E"

try:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_location)
    print(f"Successfully extracted {zip_file_path} to {extract_location}")
except FileNotFoundError:
    print(f"Error: The file {zip_file_path} was not found.")
except zipfile.BadZipFile:
    print(f"Error: The file {zip_file_path} is not a valid ZIP file.")
except Exception as e:
    print(f"An error occurred: {e}")
```
